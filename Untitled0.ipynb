```python
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import mutual_info_classif
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
# import the necessary libraries for constructing an SVM model for HAR classification.
```

```python
X_train_path = "/content/UCI HAR/Dataset/train/X_train.txt"
y_train_path = "/content/UCI HAR/Dataset/train/y_train.txt"
# defining the test and train paths
X_test_path = "/content/UCI HAR/Dataset/test/X_test.txt"
y_test_path = "/content/UCI HAR/Dataset/test/y_test.txt"
```

```python
X_train = np.loadtxt(X_train_path)
y_train = np.loadtxt(y_train_path).astype(int).ravel()
X_test = np.loadtxt(X_test_path)
y_test = np.loadtxt(y_test_path).astype(int).ravel()
# we convert the loaded test and train data files to a numpy array for making them compatible to standard scale method
# also we convert the y_train data file to a single dimensional array by using the ravel function
# because the model expects the lables to be in a 1-d format
```

```python
# Standardization
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
# applying the sandard scaler method from the scikit library
# using Z score normalization scaling method for making the data centered around
# zero and reducing the diverse range of values (X scaled = ([X- meaan]/ Standard Deviation)
```

```python
from sklearn.svm import SVC

# Initialize and train the model
svm_model = SVC(kernel='rbf', C=1, gamma='scale')  # RBF kernel is commonly used for HAR tasks
svm_model.fit(X_train, y_train)
# C=1 to prevent overfitting but also let some misclassifications for a better generalization
```

```python
# Generating Predictions on test set
y_pred = svm_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
# Calculating the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"SVM Model Accuracy: {accuracy:.4f}")

```
