```python

import pandas as pd

# reading the train data files
X_train = pd.read_csv("/content/UCI_HAR/Dataset/train/X_train.txt", delim_whitespace=True, header=None)
y_train = pd.read_csv("/content/UCI_HAR/Dataset/train/y_train.txt", delim_whitespace=True, header=None)

# reading the train data files
X_test = pd.read_csv("/content/UCI_HAR/Dataset/test/X_test.txt", delim_whitespace=True, header=None)
y_test = pd.read_csv("/content/UCI_HAR/Dataset/test/y_test.txt", delim_whitespace=True, header=None)

print(X_train.shape, y_train.shape)  # Checkingthe shape of both test and training data
```

```python
print(X_train.head())
print(y_train.head())
# checking the first few ;lines of the dataset to understand it
# X_train contains the features or the numeric values which need to be predicted for X_test
# y_train contains the labels or categories for which the values of X_train exist
```

```python
print(X_train.isnull().sum().sum(), y_train.isnull().sum().sum())
# checking for the total numbers of null values present in dataset to maintain the integrity of predictions
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)  # using the standard scalar function  from preprocessing package of sklearn
# library to scale the data
# using Z score normalization scaling method for making the data centered around
# zero and reducing the diverse range of values (X scaled = ([X- meaan]/ Standard Deviation)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100,max_depth=10,min_samples_leaf=5 ,random_state=42)
rf_model.fit(X_train_scaled, y_train.values.ravel())
# creating a random forest classifier model with 100 decision trees and random state 42(as per general convention)
#  for consistent randomness
# each time the model is ran by the user
# the max deapth of each decision tree set to 10 to prevent overfitting minimum data points for each leaf node set to
# 5 to prevent overfitting
```

```python
y_pred_rf = rf_model.predict(X_test_scaled)
# print(y_pred_rf)
# generating and storing predictions fo the y labels on the trained random forest model using the scaled data on X_test
# on basis of train data of the y labels for the values would be predicted
```

```python
from sklearn.metrics import accuracy_score

rf_accuracy = accuracy_score(y_test, y_pred_rf)  # Compare predictions with actual values

print(f"Random Forest Accuracy: {rf_accuracy:.4f}")  # Display accuracy
# cheking the accuracy of the model by comparing the y_test andy_pred_rf values and checking the prportion of
# data points where thepredicted values are same as the actual values
```
